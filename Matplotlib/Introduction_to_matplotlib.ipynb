```python
car_sales=pd.read_csv('car-sales.csv')
car_sales.rename(columns={'Odometer (KM)': 'Odometer(km)'},inplace=True)
car_sales
```

```python
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '',regex=True)
car_sales
```

```python
type(car_sales['Price'][0])
```

```python
#Remove last two zeroes
car_sales['Price'] = car_sales['Price'].str[:-2]
car_sales
```

```python
car_sales['Sales Date'] = pd.date_range('4/8/2023',periods=len(car_sales))
car_sales
```

```python
car_sales['Total Sales']=car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
#Let's plot total sales
car_sales.plot(x='Sales Date', y='Total Sales');
```

```python
car_sales['Price']=car_sales['Price'].astype(int)

car_sales.plot(x='Odometer(km)',y='Price',kind='scatter')
```
